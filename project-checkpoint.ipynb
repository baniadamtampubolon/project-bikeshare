### Mencari jumlah peminjaman sepeda dari tiap musim

```python
# Membuat sebuah function untuk menghitung jumlah peminjam sepeda dari tiap musim

def count_seasonal_user(dataframe, season):
    seasonal = day_df[day_df['season'] == season]
    
    total_casual_users = seasonal['casual'].sum()
    total_registered_users = seasonal['registered'].sum()
    total_users = total_casual_users + total_registered_users
    
    return total_users
```

```python
# Memanggil fungsi
total_spring_users = count_seasonal_user(day_df, 1)
total_summer_users = count_seasonal_user(day_df, 2)
total_fall_users = count_seasonal_user(day_df, 3)
total_winter_users = count_seasonal_user(day_df, 4)
total_users = total_spring_users + total_summer_users + total_fall_users + total_winter_users
    
# Menampilkan hasil
print("Total pengguna musim semi:", total_spring_users, "orang")
print("Total pengguna musim panas:", total_summer_users, "orang")
print("Total pengguna musim gugur:", total_fall_users, "orang")
print("Total pengguna musim salju:", total_winter_users, "orang")
print("Total pengguna = ",total_users, "orang") 
```

```python
# Membuat fungsi untuk menampilkan data melalui persentase (%)

def hitung_persentase(jumlah, total):
    return (jumlah / total) * 100
```

```python
# Memanggil Fungsi
persentase_spring = hitung_persentase(total_spring_users, total_users)
persentase_summer = hitung_persentase(total_summer_users, total_users)
persentase_fall = hitung_persentase(total_fall_users, total_users)
persentase_winter = hitung_persentase(total_winter_users, total_users)

# Menampilkan persentase hasil
print("Persentase pengguna musim semi:", round(persentase_spring, 2), "%")
print("Persentase pengguna musim panas:", round(persentase_summer, 2), "%")
print("Persentase pengguna musim gugur:", round(persentase_fall, 2), "%")
print("Persentase pengguna musim salju:", round(persentase_winter, 2), "%")
```

```python
data_2011 = day_df[day_df['yr'] == 0]  # Tahun 2011
data_2012 = day_df[day_df['yr'] == 1]  # Tahun 2012

# Mengelompokkan data berdasarkan bulan untuk setiap tahun
grouped_2011 = data_2011.groupby('mnth')['cnt'].sum()
grouped_2012 = data_2012.groupby('mnth')['cnt'].sum()

plt.plot(grouped_2011.index, grouped_2011.values, label='2011')
plt.plot(grouped_2012.index, grouped_2012.values, label='2012')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Peminjaman Sepeda')
plt.title('Peningkatan Tren Peminjaman Sepeda antara Tahun 2011 dan 2012')
plt.legend()
plt.show()
```
